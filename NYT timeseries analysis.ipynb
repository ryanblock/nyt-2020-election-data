```python
import sys
sys.path.append('./vendor')
import pandas as pd
import numpy as np
import seaborn as sns
sns.set(rc={'figure.figsize':(20, 6)})
```

```python
df = pd.read_csv("data/nyt_ts.csv", parse_dates=['timestamp'])
```

```python
df.dtypes
```

```python
df['vote_ratio_dem'] = df['vote_share_dem'] / df['vote_share_rep']
```

```python
df.plot(x='timestamp', y='votes', kind='scatter', color='blue')
```

```python
start = df['timestamp'] >= '1970-01-01'
# start = df['timestamp'] >= '2020-11-04T08'
end = df['timestamp'] < '2020-11-08'
# end = df['timestamp'] < '2020-11-05T10'
filter = df.loc[start & end]
```

```python
# a nice comparison
ca = filter[df['state'] == 'california']
ca.plot(x='timestamp', y='vote_ratio_dem', kind='scatter', color='blue')
```

```python
# another nice comparison
fl = filter[df['state'] == 'florida']
fl.plot(x='timestamp', y='vote_ratio_dem', kind='scatter', color='blue')
```
